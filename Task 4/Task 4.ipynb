DATA VISUALIZATION: We will learn how to create basic plots using matplotlib library:Scatter
plot, Histogram and Bar plot.Data visualization allows us to quickly interpret the data and adjust
different variables to see their effect.Technology is increasingly making it easier for us to do so.
Why visualize data? 1.Observe the patterns 2.Identify extreme values that could be anomalies
3.Easy interpretation.
Importing necessary libraries 1.‘pandas’ library to work with dataframes 2.‘numpy’ library to do
numerical operations 3.‘matplotlib’ library to do visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the CSV file into a pandas DataFrame
data = pd.read_csv("ToyotaCorolla.csv",na_values=["??","????"])
# Display the first few rows of the DataFrame
print("First 5 rows of the DataFrame:")
print(data.head())
# Perform some basic operations on the data
print("\nSummary statistics:")
print(data.describe())
# Visualize the data (scatter plot)
plt.scatter(data['Price'], data['Age'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title='Scatter Plot of X vs Y'
plt.show()
```

```python
plt.hist(data['KM'],color='red',edgecolor='white',bins=5)
plt.title='Histogram of KM'
plt.xlabel('KM')
plt.ylabel('frequency')
```

```python
counts = [979, 120, 12]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))
plt.bar(index,counts,color=['red','blue','cyan'])
plt.title="Bar plot of fuel types"
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index, fuelType, rotation = 90)
plt.show()
```

```python
data = pd.read_csv('ToyotaCorolla.csv')
# Scatter plot with day against tip
plt.scatter(data['Age'], data['Price'])
# Adding Title to the Plot
plt.title="Scatter Plot"
# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data['Age'], data['Price'], c=data['MetColor'],
s=data['MetColor'])
# Adding Title to the Plot
plt.title="Scatter Plot"
# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('Price')
plt.colorbar()
plt.show()
```

```python
plt.plot(data['Age'])
plt.plot(data['Price'])
# Adding Title to the Plot
plt.title="Scatter Plot"
# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('price')
plt.show()
```
